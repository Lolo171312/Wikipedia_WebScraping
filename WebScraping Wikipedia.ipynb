```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Save url as variable
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

#Request page
page = requests.get(url, headers = headers)

#Parser the requested page
soup = BeautifulSoup(page.content, "html.parser")
```

```python
#Get table´s title through the id
table_title = soup.find(id = "List_of_the_largest_public_/_publicly_traded_companies").get_text().strip()
table_title
```

```python
#Get table content. I found a "table" section with the class specifier because there are more than two tables
table_content = soup.find('table', class_="wikitable sortable")
```

```python
#Get columns titles
column_titles = table_content.find_all("th")
fixed_column_titles = [title.get_text().strip() for title in column_titles]
fixed_column_titles
```

```python
#Create column titles in a pandas DataFrame based on the previously found column_titles
df = pd.DataFrame(columns = fixed_column_titles)
df
```

```python
#Get every data in the table
table_data = table_content.find_all("tr")
row_index = 0
#Get data for every specific row
#The first index of the list is Null so I start the loop with the index 1
for row in table_data[1:]:
    row_data = [data.get_text().strip() for data in row.find_all("td")]

    #Insert found data to the DataFrame
    df.loc[row_index] = row_data
    row_index += 1
```
